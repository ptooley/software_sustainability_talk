```python
import numpy as np

from IPython.display import Image, Latex, display
%matplotlib inline
%config InlineBackend.print_figure_kwargs = {'bbox_inches':None}
import matplotlib.pyplot as plt
from matplotlib import rcParams
import matplotlib.ticker as mt
import matplotlib.cm as mcm
import matplotlib.lines as mlines

import pyutils.matplotlib as pu_m
from pyutils.ipython import ProgressBar

rcParams['text.usetex'] = False
```

```python
nax = 2
lpad = 0.05
rpad = 0.05
ipad = 0.05
width = (1-lpad - rpad - (nax-1)*ipad)/nax
tpad = 0.15
bpad = 0.05
height = 1 - tpad - bpad
axlayout = [(lpad + i*(ipad+width),bpad, width, height) for i in range(nax)]
fig = plt.figure(figsize=(8,8/nax), facecolor='grey')
ax = [fig.add_axes(fp) for fp in axlayout]

colors = [mcm.tab20c.colors[0],
          mcm.tab20c.colors[4]]

pieargs = {'radius': 1,
           'wedgeprops': {'width':0.45},
           'startangle': 90,
           'counterclock': False,
           'colors': colors,
           'labeldistance': 0.75,
           'textprops': {'ha': 'center'}}

ax[0].pie([92, 8], labels=['Yes\n(92%)', 'No\n(8%)'], **pieargs)
ax[1].pie([56, 44], labels=['Yes\n(56%)', 'No\n(44%)'], **pieargs)

ax[0].set_title('Do you use research software?')
ax[1].set_title('Do you develop your own\nresearch software?')

display(fig)
fig.savefig('surveyres.png',dpi=300)
plt.close(fig)
```
